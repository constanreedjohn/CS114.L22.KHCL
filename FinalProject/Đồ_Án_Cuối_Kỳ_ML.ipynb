# **ĐỒ ÁN CUỐI KỲ:** Dự đoán điểm chuẩn của các trường đại học.

# **1. Mô tả bài toán:**
> Hiện nay, các trường đại học thường công bố kết quả điểm chuẩn sau khi thí sinh thi xong, việc chờ đợi kết quả sẽ gây lại lo lắng cho các thí sinh, và sẽ ảnh hưởng tới việc đăng ký nguyện vọng cho phù hợp.

> Vì thế tụi em muốn sử dụng các thuật toán Machine Learing để có thể dự đoán được điểm chuẩn của các trường đại học.

> **Ứng dụng** của việc sẽ giúp cho các thí sinh có thể dựa vào đó và đăng ký nguyện vọng vừa sức mình mà không bị chọn các nguyện vọng không vừa ý.

> **Ngữ cảnh:** việc dự đoán điểm chuẩn chỉ mới thực hiện bởi các thầy có kinh nghiệm trong lĩnh vực xét về chỉ tiêu, độ hot của ngành... Tụi em muốn đưa máy học vào để có thể tăng độ chính xác hơn.

> **INPUT:** Tập dữ liệu dataset với các thông tin về điểm chuẩn.

> **OUTPUT:** Điểm chuẩn của các trường đại học.

# **2. Mô tả về dataset:**

> Dataset của tụi em sẽ gồm có các thông tin:
* Mã trường: Mã trường đại học
* Tên trường: Tên trường đại học
* Mã ngành: Mã ngành của trường
* Tên ngành: Tên ngành của trường
* Năm: Năm xét tuyển
* Số chỉ tiêu: Số chỉ tiêu theo năm
* Phương thức xét tuyển: gồm "Chính Qui", "Học Bạ".
* Điểm chuẩn: Điểm chuẩn qua các ngành theo từng năm.

> **Cách thức xây dựng dataset:**
>> Về cách xây dựng dataset thì tụi em sẽ crawl data trên các trang web trường đại học hoặc các trang báo chí.

> **Số lượng, độ đa dạng:**
>> Số lượng dataset sẽ khoảng 3000 dữ liệu, split train/test theo 80/20.
>> Độ đa dạng sẽ tập trung chủ yếu ở các năm xét tuyển, và các ngành.
